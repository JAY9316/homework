```python
import matplotlib.pyplot as plt
import numpy as np

# 第一組條件
a1, b1, c1 = -2, 4, 5

# x值範圍
x = np.arange(1, 501)

# 計算第一組條件下的y值
y1 = a1 * x ** 2 + b1 * x + c1

# 找出最高點和最低點的索引
max_index = np.argmax(y1)
min_index = np.argmin(y1)

# 最高點的座標
max_point = (x[max_index], y1[max_index])
# 最低點的座標
min_point = (x[min_index], y1[min_index])

# 繪製xy軸圖
plt.plot(x, y1, label='Group 1: a=-2, b=4, c=5')

# 標註最高點和最低點的座標值
plt.scatter(max_point[0], max_point[1], color='red')
plt.scatter(min_point[0], min_point[1], color='blue')
plt.annotate(f'Max: {max_point}', xy=max_point, xytext=(max_point[0], max_point[1] - 10), color='red')
plt.annotate(f'Min: {min_point}', xy=min_point, xytext=(min_point[0], min_point[1] + 10), color='blue')

# 加上十字XY軸
plt.axhline(0, color='black')
plt.axvline(0, color='black')

# 加上標籤和圖例
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# 顯示圖形
plt.show()
```
